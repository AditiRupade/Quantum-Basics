Qiskit is an open source SDK for working with IBM Quantum processor. You can use it locally by consuming multiple APIs provided by [Qiskit](https://qiskit.org/documentation/getting_started.html) or you can also run it on [IBM Quantum Cloud](https://quantum-computing.ibm.com/).

## Quantum State Simulation

Qiskit provides a way to draw the state of the quantum circuit.

```python
from qiskit.quantum_info import Statevector
from qiskit import QuantumCircuit

quantum_circuit = QuantumCircuit(1)
quantum_circuit.x(0)
quantum_circuit.draw('mpl')
```

```python
state = Statevector.from_int(0, 2**1)
state = state.evolve(quantum_circuit)
state.draw('latex')
```

```python
from qiskit.visualization import array_to_latex

#Alternative way of representing in latex
array_to_latex(state)
```

```python
state.draw('qsphere')
```

```python
state.draw('hinton')
```

## Unitary Respresentation

```python
from qiskit.quantum_info import Operator

U = Operator(quantum_circuit)
U.data
```

## Measuring Qubit

```python
quantum_circuit = QuantumCircuit(1,1)
quantum_circuit.h(0)
quantum_circuit.measure(0,0)

quantum_circuit.draw('mpl')
```

```python
from qiskit.providers.aer import QasmSimulator
from qiskit import transpile

qc_compiled = transpile(quantum_circuit, QasmSimulator())
```

```python
job = QasmSimulator().run(qc_compiled, shots=1024)
results = job.result()
counts = results.get_counts(qc_compiled)
print(counts)
```

## Quantum Visualization

```python
from qiskit.visualization import plot_histogram
plot_histogram(counts)
```

```python
from qiskit.visualization import plot_bloch_vector
plot_bloch_vector([0,0,1])
```
